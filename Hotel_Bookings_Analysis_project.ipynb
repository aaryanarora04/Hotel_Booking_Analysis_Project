#Data Analysis & Visualisations

```python
df['is_canceled'].value_counts(normalize=True)
```

```python
plt.figure(figsize=(5,4))
plt.bar(['Not Canceled', 'Canceled'], df['is_canceled'].value_counts(), edgecolor='k', width=0.5)
plt.title("Reservation Status Count")
plt.xlabel('Status')
```

```python
#plt.figure(figsize=(5,4))
sns.countplot(x='hotel', data= df, hue='is_canceled', palette='Blues')
plt.title("Reservation in different hotels", size=20)
plt.xlabel("Hotel")
plt.ylabel("No. of Reservations")
```

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20,7))
plt.title("Avg Daily Rate in City and Resort Hotel", fontsize=20)
plt.plot(resort_hotel.index, resort_hotel['adr'])
plt.plot(city_hotel.index, city_hotel['adr'])
plt.legend(['Resort Hotel', 'City Hotel'], fontsize = 15)
```

```python
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize=(10,6))
sns.countplot(x= df['month'], data=df, hue= df['is_canceled'])
```

```python
plt.figure(figsize=(10,6))
plt.title('ADR per month', fontsize=15)
sns.barplot(x= 'month', y='adr', data= df[df['is_canceled'] == 1].groupby('month')['adr'].sum().reset_index())
```

```python
canceled_reservations = df[df['is_canceled'] == 1] 

top10_countries = canceled_reservations['country'].value_counts()[:10]

plt.figure(figsize=(8,6))
plt.title("Top 10 Countries with Reservation Canceled")
plt.pie(top10_countries, autopct="%.2f", labels = top10_countries.index)
plt.show()
```

```python
df['market_segment'].value_counts()
```

```python
df['market_segment'].value_counts(normalize=True)
```

```python
canceled_reservations['market_segment'].value_counts(normalize=True)
```

```python
canceled_reservations_adr = canceled_reservations.groupby('reservation_status_date')[['adr']].mean()
canceled_reservations_adr.reset_index(inplace=True)
canceled_reservations_adr.sort_values('reservation_status_date', inplace=True)

non_canceled_reservations = df[df['is_canceled'] == 0]
non_canceled_reservations_adr = non_canceled_reservations.groupby('reservation_status_date')[['adr']].mean()
non_canceled_reservations_adr.reset_index(inplace=True)
non_canceled_reservations_adr.sort_values('reservation_status_date', inplace=True)

plt.figure(figsize=(20,6))
plt.title("Avg Daily Rate", fontsize=20)
plt.plot(canceled_reservations_adr['reservation_status_date'], canceled_reservations_adr['adr'], label="Canceled" )
plt.plot(non_canceled_reservations_adr['reservation_status_date'], non_canceled_reservations_adr['adr'], label="Non Canceled" )
plt.legend(fontsize=15)
plt.show()
```

```python
canceled_reservations_adr = canceled_reservations_adr[(canceled_reservations_adr['reservation_status_date'] >'2016') & (canceled_reservations_adr['reservation_status_date']<'2017-09')]
non_canceled_reservations_adr = non_canceled_reservations_adr[(non_canceled_reservations_adr['reservation_status_date'] >='2016') & (non_canceled_reservations_adr['reservation_status_date']<'2017-09')]
```

```python
plt.figure(figsize=(20,6))
plt.title("Avg Daily Rate", fontsize=20)
plt.plot(canceled_reservations_adr['reservation_status_date'], canceled_reservations_adr['adr'], label="Canceled" )
plt.plot(non_canceled_reservations_adr['reservation_status_date'], non_canceled_reservations_adr['adr'], label="Non Canceled" )
plt.legend(fontsize=15)
plt.show()
```
